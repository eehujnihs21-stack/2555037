```python
!pip install -q gradio

import gradio as gr
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif

# =====================================================================
# 1. 데이터 로드 및 전처리 (Feature Selection)
# =====================================================================
# 와인 데이터셋 불러오기
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target

# 통계적 기여도가 가장 높은 핵심 특징(Feature) 3개 자동 선택
selector = SelectKBest(score_func=f_classif, k=3)
X_selected = selector.fit_transform(X, y)

# 선택된 3개 특징의 이름 추출
selected_features = X.columns[selector.get_support()].tolist()
print(f"✅ 선택된 3개의 핵심 특징: {selected_features}")

# =====================================================================
# 2. AI 예측 모델 학습 (Single Server 내장용)
# =====================================================================
# 분류 성능이 뛰어난 랜덤 포레스트(RandomForest) 모델 학습
model = RandomForestClassifier(random_state=42)
model.fit(X_selected, y)

# =====================================================================
# 3. Gradio 웹 인터페이스 및 함수 정의
# =====================================================================
# 웹 화면에서 입력받은 값을 모델에 넣어 예측하는 함수
def predict_wine(feature_1, feature_2, feature_3):
    # 입력 데이터를 모델 형식에 맞게 DataFrame으로 변환
    input_df = pd.DataFrame([[feature_1, feature_2, feature_3]], columns=selected_features)

    # 모델 예측 수행 (0, 1, 2 중 하나 반환)
    prediction = model.predict(input_df)[0]

    # 숫자로 된 예측 결과를 실제 와인 품종 이름으로 변환하여 반환
    wine_name = wine.target_names[prediction].upper()
    return f"이 와인은 【 {wine_name} 】 품종으로 예측됩니다."

# Gradio UI 레이아웃 구성
demo = gr.Interface(
    fn=predict_wine,
    inputs=[
        gr.Number(label=f"1. {selected_features[0]}"),
        gr.Number(label=f"2. {selected_features[1]}"),
        gr.Number(label=f"3. {selected_features[2]}"),
    ],
    outputs=gr.Textbox(label="인공지능 예측 결과"),
    title="🍷 와인 품종 자동 예측 서비스",
    description=f"와인의 3가지 핵심 화학 성분 값을 입력하면 인공지능이 와인 품종을 분류해 줍니다.",
)

# =====================================================================
# 4. 싱글 서버 구동 (Gradio 마운트)
# =====================================================================
if __name__ == "__main__":
    demo.launch(share=True)
```
